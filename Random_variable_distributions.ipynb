## Labs Random Variable Distributions

```python
import pandas as pd
import numpy as np
```

1. Get the numerical variables from our dataset.

```python
customers = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
customers.head()
```

```python
customers.rename(columns=lambda x: x.strip().lower().replace(" ","_"), inplace=True)
```

```python
numerical_data = customers.select_dtypes(include=np.number)
numerical_data
```

2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
sns.distplot(numerical_data['total_claim_amount'], fit=norm)
plt.show()
```

```python
from scipy.stats import expon

sns.distplot(numerical_data['total_claim_amount'], fit=norm)
plt.show()
```
